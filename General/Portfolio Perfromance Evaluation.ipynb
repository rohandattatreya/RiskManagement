```python
# Dollar weighted return

def dollar_weighted_return(cashflows, beginningbalance, endingbalance):
    """
    Calculate the dollar weighted return
    """
    cashflows.insert(0, -beginningbalance)
    cashflows.append(endingbalance)
    cashflows = [cashflows[i] / cashflows[i - 1] for i in range(1, len(cashflows))]
    dollar_weighted_return = 1
    for cashflow in cashflows:
        dollar_weighted_return *= cashflow
    return dollar_weighted_return - 1
```

```python
# Sharpe Ratio
def sharpe_ratio(portfolio_return, risk_free_rate, portfolio_volatility):
    """
    Calculate the Sharpe ratio
    """
    sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_volatility
    return sharpe_ratio
```

```python
# Sortino Ratio
def downside_risk(portfolio_returns, target):
    """
    Calculate the downside risk
    """
    downside_returns = [return_ for return_ in portfolio_returns if return_ < target]
    downside_risk = sum([return_ ** 2 for return_ in downside_returns]) / len(portfolio_returns)
    return downside_risk ** 0.5

def sortino_ratio(portfolio_return, risk_free_rate, downside_risk):
    """
    Calculate the Sortino ratio
    """
    sortino_ratio = (portfolio_return - risk_free_rate) / downside_risk
    return sortino_ratio
```

```python
# Treynor Ratio
def treynor_ratio(portfolio_return, risk_free_rate, beta):
    """
    Calculate the Treynor ratio
    """
    treynor_ratio = (portfolio_return - risk_free_rate) / beta
    return treynor_ratio
```

```python
# Information Ratio
## Higher the information ratio, the better the performance of the portfolio

def information_ratio(average_return, benchmark_return, tracking_error):
    return (average_return - benchmark_return) / tracking_error
```

# M-squared 
## higher the M-squared, the better the performance of the portfolio
## M-squared is a risk-adjusted performance measure that is used to evaluate the performance of a portfolio.
Taking into account the risk-free rate of return and the benchmark return. 
It is calculated by adding the product of the tracking error and the difference between the average return and the benchmark return to the average return minus the risk-free rate of return.




```python
# M-squared 

def m_squared(average_return, benchmark_return, tracking_error, risk_free_rate):
    return (average_return - risk_free_rate) + (tracking_error * (average_return - benchmark_return))

    
```

```python
# statistical significance of alpha
def alpha_significance(alpha, tracking_error, n):
    """
    Calculate the statistical significance of alpha
    alpha: float, alpha of the portfolio
    tracking_error: float, standard error of alpha
    n: int, number of observations
    
    
    """
    t_stat = round(alpha / (tracking_error / (n ** 0.5)),2)
    return t_stat
```
